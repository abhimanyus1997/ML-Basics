# Linked List

### Node Class:

```python
class Node:
    def __init__(self,value) -> None:
        self.value = value
        self.next = None    # intially we dont know it it has neighbour
```

```python
new_node = Node(10)
print(new_node)
```

* Time complexity of Assignment & initilzation : O(1)
* Space complexity of Assignment & initilzation : O(1)

## Singly Linked List

```python
class LinkedList:
    def __init__(self, value) -> None:
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
```
