# Linked List

### Node Class:

```python
class Node:
    def __init__(self,value) -> None:
        self.value = value
        self.next = None    # intially we dont know it it has neighbour
```

```python
new_node = Node(10)
print(new_node)
```

* Time complexity of Assignment & initilzation : O(1)
* Space complexity of Assignment & initilzation : O(1)

## Singly Linked List

```python
class LinkedList:
    def __init__(self, value) -> None:
        new_node = Node(value)          #O(1)
        self.head = new_node            #O(1)
        self.tail = new_node            #O(1)
        self.length = 1                 #O(1)

# Time complexity of Assignment & initilzation : O(1)
# Space complexity of Assignment & initilzation : O(1)
```

```python
new_linked_list = LinkedList(value=10)

# Print the details of the linked list
print("New Linked List:", new_linked_list)
print("Head value:", new_linked_list.head.value)
print("Length of the linked list:", new_linked_list.length)
```

### Insertion to Linked List

#### At the End

* create new node
* update pointer of last node from `None` to new node



> Edge Cases: If no element is there, Head and Tail should point to New Node 

```python
class LinkedList:
    def __init__(self) -> None:
        # Initialize the linked list with head, tail, and length attributes
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, value):
        # To check if the node is null or not
        # Method 1 : Head attribute points to None
        # Method 2 : Length Attribute == 0

        # Step 1 : Create Node
        new_node = Node(value)

        # Edge Case: If the linked list is empty
        if self.head is None:
            # Step 2: Update Head^Tail to new node
            self.head = new_node
            self.tail = new_node

        else:
            # Step 2: Append the new node to the end of the linked list
            self.tail.next = new_node
            self.tail = new_node    # Update the tail to the new node
        self.length += 1  # increase length


    def __str__(self) -> str:
        temp_node = self.head
        result = ''
        while temp_node is not None:
            result += str(temp_node.value)
            if temp_node.next is not None:
                result += '->'
            temp_node = temp_node.next
        return result
```

```python
new_linked_list = LinkedList()
new_linked_list.append(10)
new_linked_list.append(20)

# Visulaization of Node
print(f"Singly Linked List:\n{new_linked_list}\n")
print("No of Nodes:",new_linked_list.length)
```
